# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

# File to Load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_type_count = pyber_data_df.groupby(['type']).count()['ride_id']
ride_type_count
```

```python
# Build total rides for each city type pie chart
plt.subplots(figsize=(10, 6))
plt.pie(ride_type_count,
    labels=["Rural", "Suburban", "Urban"],
    colors=["lavender", "lightsteelblue", "royalblue"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Change the default font size from 10 to 16.
mpl.rcParams['font.size'] = 16

# Save Figure
plt.savefig("Analysis/Pyber_TotalRides_CityType")

# Show Figure
plt.show()
```

```python
# 2. Get the total drivers for each city type
driver_type_count = city_data_df.groupby(['type']).sum()['driver_count']
driver_type_count
```

```python
# Build percentage of rides by city type pie chart.
plt.subplots(figsize=(10, 6))
plt.pie(driver_type_count,
    labels=["Rural", "Suburban", "Urban"],
    colors=["lavender", "lightsteelblue", "royalblue"],
    explode=[0, 0, 0.1],
    autopct='%1.1f%%',
    shadow=True, startangle=165)
plt.title("% of Total Drivers by City Type")

# Change the default font size from 10 to 16.
mpl.rcParams['font.size'] = 16

# Save Figure
plt.savefig("Analysis/Pyber_TotalDrivers_CityType")

# Show Figure
plt.show()
```

```python
#  3. Get the total amount of fares for each city type
fares_type_total = pyber_data_df.groupby(['type']).sum()['fare']
fares_type_total
```

```python
#  4. Get the average fare per ride for each city type. 
fare_type_mean = pyber_data_df.groupby(['type']).mean()['fare']
fare_type_mean
```

```python
# 5. Get the average fare per driver for each city type. 
perdriver_fare_mean = fares_type_total / driver_type_count
perdriver_fare_mean
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    'Total Rides': ride_type_count,
    'Total Drivers': driver_type_count,
    'Total Fares': fares_type_total,
    'Average Fare per Ride': fare_type_mean,
    'Average Fare per Driver': perdriver_fare_mean
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map("{:,.0f}".format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map("{:,.0f}".format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
type_sumfares_df = pyber_data_df.groupby(['type','date']).sum()[['fare']]
type_sumfares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. 
# This is needed to use the 'pivot()' function.
type_sumfares_df = type_sumfares_df.reset_index()
type_sumfares_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
type_sumfares_pivotdf = type_sumfares_df.pivot(index = 'date', columns = 'type', values = 'fare')
type_sumfares_pivotdf
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates
# '2019-01-01':'2019-04-29'.
dates_sumfares_df = type_sumfares_pivotdf.loc['2019-01-01':'2019-04-29']
dates_sumfares_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
dates_sumfares_df.index = pd.to_datetime(dates_sumfares_df.index)
dates_sumfares_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
dates_sumfares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
dates_sumfares_week_df = dates_sumfares_df.resample('W').sum()
dates_sumfares_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

ax = dates_sumfares_week_df.plot(figsize=(17,6))
ax.plot()

# Set axis and title names
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
ax.set_xlabel('Month')

plt.savefig("Analysis/PyBer_fare_summary.png")
plt.show()
```
